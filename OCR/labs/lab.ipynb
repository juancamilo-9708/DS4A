```python
import os
import json

outfile = open('result.txt', 'w')
path = 'output_txt_v1/ouput_txt/'
files = os.listdir(path)
total_text = ''
for file in files:
    print(f'Reading file {file}')
    with open(path+file) as f:
        lines = f.readlines()
        lines=lines[0]
        lines = json.loads(lines)
        for line in lines :
            print(f'Reading {line}')
            total_text = lines[line] + total_text
    
outfile.close()
```

```python
import codecs
file = codecs.open("compiled_text", "w+", "utf-8")
file.write(total_text)
```

```python
from wordcloud import WordCloud,STOPWORDS
```

```python
total_text = total_text.replace('\n', ' ')
```

```python
import matplotlib.pyplot as plt
stopwords = set(STOPWORDS)
stopwords.update(["de", "la", "en", "el", "y", "e","por", "para", "que", "lo", "con lo", "los", "o"])
wordcloud = WordCloud(stopwords=stopwords,max_font_size=50, max_words=100, background_color="black").generate(total_text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
